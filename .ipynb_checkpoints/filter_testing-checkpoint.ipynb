```python
def count_seqs(filename):
    i = 0
    for record in SeqIO.parse(filename, "fastq"):
        i = i + 1
    return i
```

```python
def parse_fastq(seq_len,n_seqs):
    i = 0
    quality_matrix = np.zeros((seq_len,n_seqs))
    sequences = []
    for record in SeqIO.parse(filename, "fastq"):
        score=record.letter_annotations["phred_quality"]
        sequences.append(record)
        quality_matrix[:,i] = score   
        i = i +1
    return quality_matrix, sequences
```

```python
filename = "/Volumes/Storage/A1/test_r1.fastq"
n_seqs = count_seqs(filename)
seq_len = 100
```

```python
input_seq_iterator = SeqIO.parse("cor6_6.gb", "genbank")
short_seq_iterator = (record for record in input_seq_iterator if len(record.seq) < 300)
```

```python
quality_matrix_r1, sequences_r1 = parse_fastq(seq_len,n_seqs,
                                            "/Volumes/Storage/A1/test_r1.fastq")
quality_matrix_r2, sequences_r2 = parse_fastq(seq_len,n_seqs,
                                            "/Volumes/Storage/A1/test_r2.fastq")
to_retain = (np.mean(quality_matrix_r1[35:100,:],axis=0) > 30) & (np.mean(quality_matrix_r2[35:100,:],axis=0) > 30)

with open("/Volumes/Storage/A1/filtered_r1.fastq", "w") as output_handle:
    SeqIO.write(list(compress(sequences_r1, to_retain)), output_handle, "fastq")
with open("/Volumes/Storage/A1/filtered_r2.fastq", "w") as output_handle:
    SeqIO.write(list(compress(sequences_r2, to_retain)), output_handle, "fastq")
```
